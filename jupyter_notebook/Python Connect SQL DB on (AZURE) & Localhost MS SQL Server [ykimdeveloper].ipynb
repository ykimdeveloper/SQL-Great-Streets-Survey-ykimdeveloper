# Geo Table

```python
df4 = df[['ID','Location',
       'Location Between', 'Location Between AND',
       'Side of street for this interview',
        'lat', 'lng', 'frmt_addrss', 'zipcode',
       'area', 'lat_lng_geotype']].copy()

#cannot change float column of values with nan into string
df4['zipcode'].fillna(0, inplace=True) #change nan values to 0
df4['zipcode'] = df4['zipcode'].apply(int) #change zipcode float values to int
df4['zipcode']= df4['zipcode'].apply(str) #change zipcode str 
df4['zipcode'] = df4['zipcode'].replace('0', 'none')

df4.to_sql("Geo", engine)
 
df4.to_sql("Geo", engine_azure)
```
